```python
import pandas as pd
```

```python
df = pd.concat(
    map(pd.read_csv, ["/content/Faculty_Asia.csv", "/content/Faculty_aus.csv", "/content/Faculty_eu.csv", "/content/Faculty_na.csv", "/content/Faculty_sa.csv"]), ignore_index=True)

df
```

```python
orcid = df["ORCID"].values.tolist()
print(orcid)
```

```python
orcid[1][:4]
```

```python
for i in orcid:
  try:
    if i[:4]!= '0000':
      df['ORCID'] = df['ORCID'].replace(i, "")
  except:
    continue
```

```python
df
```

```python
df.to_csv("Faculty.csv", index=False)
```
